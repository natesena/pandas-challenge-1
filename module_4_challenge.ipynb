## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df["subtotal"] = df["qty"] * df["unit_price"]
df.head(5)
```

```python
# add weight column
df["total_weight"] = df["qty"]*df["unit_weight"]
df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
def calculate_shipping(qty, unit_weight):
    weight = qty * unit_weight
    return weight * 7 if weight > 50 else weight * 10

df['shipping_price'] = df.apply(lambda row: calculate_shipping(row['qty'], row['unit_weight']), axis=1)
df.head()
```

```python
# check work
less_than_fifty_lbs = df.loc[df["shipping_price"] < 350]
less_than_fifty_lbs
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
def calculate_total_price(subtotal, shipping):
    return round((subtotal+shipping)*1.0925, 2)
    
df["total_price"] = df.apply(lambda row: calculate_total_price(row['subtotal'], row['shipping_price']), axis=1)
df.head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
def total_unit_cost(unit_cost, qty, shipping):
    if qty == 0:
        unit_shipping=0
    else:
        unit_shipping = shipping/qty
    return unit_cost + unit_shipping

df["total_unit_cost"]= df.apply(lambda row: total_unit_cost(row['unit_cost'], row['qty'], row['shipping_price']), axis=1)
df.head(50)
```

```python
# Create a column for the profit of
# each line using line cost and line price
df["line_profit"] = ((df["unit_price"] - df["unit_cost"])*df["qty"])
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
total_client_quantity = df.groupby('client_id').agg({'qty': 'sum', 'total_price': 'sum'}).reset_index()
total_client_quantity_sorted = total_client_quantity.sort_values(by='qty', ascending=False)
total_client_quantity_sorted.head(5)
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
summary_df = df.groupby('client_id').agg({'qty': 'sum', 'shipping_price': 'sum', 'total_price': 'sum', 'line_profit':'sum'}).reset_index()
summary_df_sorted = summary_df.sort_values(by='line_profit', ascending=False)
summary_df_sorted.head(5)
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
summary_df_sorted = summary_df_sorted.rename(columns={'shipping_price':"total_shipping_cost", "line_profit":"total_profit"})
summary_df_sorted.head(5)
```
